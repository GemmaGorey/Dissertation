```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
```

```python
def plot_spectrogram_and_save(signal, sample_rate, output_path='spectrogram.png'):
    """
    Generate a spectrogram from audio data and save it as an image
    """
    # Create plt figure
    plt.figure(figsize=(12, 6))
    
    # Create the spectrogram
    D = librosa.amplitude_to_db(np.abs(librosa.stft(signal)), ref=np.max)
    
    # Display the spectrogram
    librosa.display.specshow(D, sr=sample_rate, x_axis='time', y_axis='log')
    
    # Add colorbar and title
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram')
    plt.tight_layout()
    
    # Create the output directory if it doesn't exist
    output_dir = os.path.dirname(output_path)
    if output_dir and not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Save the figure
    plt.savefig(output_path)
    print(f"Spectrogram saved to {output_path}")
    
    # Display the plot
    plt.show()
```

```python
def main():
    # Use the correct file path with librosa
    file_path = os.path.join('..', 'data', 'DEAM archive', 'DEAM_audio', 'MEMD_audio', '2.mp3')
    
    # Load the audio file using librosa
    try:
        signal, sample_rate = librosa.load(file_path)
        print(f'Sample rate: {sample_rate}')
        print(f'Duration: {librosa.get_duration(y=signal, sr=sample_rate):.2f} seconds')
        
        # Create and save spectrogram
        img_path = os.path.join('img', 'spectrogram.png')
        plot_spectrogram_and_save(signal, sample_rate, img_path)
        
    except Exception as e:
        print(f"Error loading audio file: {e}")
        print(f"Checking if file exists: {os.path.exists(file_path)}")
        print(f"Current working directory: {os.getcwd()}")

```

```python
if __name__ == "__main__":
    main()
```
